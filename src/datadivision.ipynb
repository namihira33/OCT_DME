```python
import numpy as np
from glob import glob
import config
import os
from PIL import Image

SAVE_N = 8616

image_path_dir = config.train_path

image_DME_paths = np.random.choice(glob(os.path.join(image_path_dir,"DME/*.jpeg")),SAVE_N,replace=False)
image_DME_movedir = os.path.join(image_path_dir,"DME_under")

image_NORMAL_paths = np.random.choice(glob(os.path.join(image_path_dir,"NORMAL/*.jpeg")),SAVE_N,replace=False)
image_NORMAL_movedir = os.path.join(image_path_dir,"NORMAL_under")

image_CNV_paths = np.random.choice(glob(os.path.join(image_path_dir,"CNV/*.jpeg")),SAVE_N,replace=False)
image_CNV_movedir = os.path.join(image_path_dir,"CNV_under")

image_DRUSEN_paths = np.random.choice(glob(os.path.join(image_path_dir,"DRUSEN/*.jpeg")),SAVE_N,replace=False)
image_DRUSEN_movedir = os.path.join(image_path_dir,"DRUSEN_under")


for image_CNV_path in image_CNV_paths:
    image_CNV = Image.open(image_CNV_path).convert('L')
    image_CNV_name = image_CNV_path.split('/')[-1].split('\\')[-1]
    delete_path = os.path.join(image_path_dir,"CNV",image_CNV_name)
    print(image_CNV_name,delete_path)
    save_path = os.path.join(image_CNV_movedir,image_CNV_name)
    image_CNV.save(save_path)
    os.remove(delete_path)

```

```python
for image_DME_path in image_DME_paths:
    image_DME = Image.open(image_DME_path).convert('L')
    image_DME_name = image_DME_path.split('/')[-1].split('\\')[-1]
    print(image_DME_name)
    delete_path = os.path.join(image_path_dir,"DME",image_DME_name)
    save_path = os.path.join(image_DME_movedir,image_DME_name)
    image_DME.save(save_path)
    os.remove(delete_path)

```

```python
for image_NORMAL_path in image_NORMAL_paths:
    image_NORMAL = Image.open(image_NORMAL_path).convert('L')
    image_NORMAL_name = image_NORMAL_path.split('/')[-1].split('\\')[-1]
    print(image_NORMAL_name)
    delete_path = os.path.join(image_path_dir,"NORMAL",image_NORMAL_name)
    save_path = os.path.join(image_NORMAL_movedir,image_NORMAL_name)
    image_NORMAL.save(save_path)
    os.remove(delete_path)
```

```python
for image_DRUSEN_path in image_DRUSEN_paths:
    image_DRUSEN = Image.open(image_DRUSEN_path).convert('L')
    image_DRUSEN_name = image_DRUSEN_path.split('/')[-1].split('\\')[-1]
    print(image_DRUSEN_name)
    delete_path = os.path.join(image_path_dir,"DRUSEN",image_DRUSEN_name)
    save_path = os.path.join(image_DRUSEN_movedir,image_DRUSEN_name)
    image_DRUSEN.save(save_path)
    os.remove(delete_path)
```
